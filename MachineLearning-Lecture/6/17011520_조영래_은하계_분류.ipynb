## 하이퍼파라미터 조정해보기

````python
def search_params(model, params, metric, X, y, verbose=True):
  print(f'<<<{model.__class__}>>>')
  
  score_with_params = []
  param_names = params.keys()
  X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                      stratify=y,
                                                      random_state=RANDOM_STATE)

  for param in product(*params.values()):
    param_dict = dict(zip(param_names, param))
    model.set_params(**param_dict)
    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_metric = metric(y_train, y_train_pred)
    test_metric = metric(y_test, y_test_pred)
    
    if verbose:
      print(f'##### {param_dict} #####')
      print(f'train {metric.__class__}: {train_metric}')
      print(f'test {metric.__class__}: {test_metric}')
      print('\n')

    score_with_params.append((train_metric, test_metric, param_dict))

  return score_with_params
````

````python
params = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['saga']
}
````

````python
accuracy_with_params = search_params(LogisticRegression(), params, accuracy_score,
                                     X_train_std, y_train, verbose=False)
````

````python
for accuracy_params in sorted(accuracy_with_params,
                              reverse=True,
                              key= lambda x: x[1])[:10]:
  print(accuracy_params)
````

````python
# submission 만드는 함수
# params : (n_parameters, )의 형태로 dict 형태로 parameter 들어있음

def submission_by_params(model, params, X, y, test, submission, file_name_prefix='submission'):
  for param in params:
    model.set_params(**param)
    model.fit(X, y)
    
    y_pred = model.predict(test)
    y_pred_inv = le_y.inverse_transform(y_pred)
    sample_submit['Category'] = y_pred_inv
    
    file_name = f'{file_name_prefix}_{param}.csv'
    sample_submit.to_csv(file_name, index=False)
````

````python
submission_params = [x[2] for x in sorted(accuracy_with_params, 
                                          reverse=True,
                                          key= lambda x: x[1])[:10]]
````

````python
submission_by_params(LogisticRegression(),
                     submission_params,
                     X_train_std, y_train, X_test_std, 
                     sample_submit,
                     file_name_prefix='std_submission')
````
